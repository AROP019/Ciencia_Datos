**3. What is the percentage of German citizens (*native-country* feature)?**

```python
# You code here
ciudadanos = data['age'].count()


ciudadanos_alemanes = data[data['native-country'] == 'Germany']
ciudadanos_alemanes_count = ciudadanos_alemanes['age'].count()

por_alemanes = (ciudadanos_alemanes_count/ciudadanos)*100
print("Entre toda los individuos registrados, el: ", por_alemanes, "son de ciudadania alemana")
```

4-5. __What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?__

```python
# You code here
mas_50 = data[data['salary'] == '>50K']
menos_50 = data[data['salary'] == '<=50K']

mas_50_edad = mas_50['age']
menos_50_edad = menos_50['age']

mas_50_edad_mean = mas_50_edad.mean()
menos_50_edad_mean = menos_50_edad.mean()

mas_50_edad_std = mas_50_edad.std()
menos_50_edad_std = menos_50_edad.std()

print("Para los que ganan más de 50k al año, el promedio de edad es: ", mas_50_edad_mean, "y la desviación estándar es: ", mas_50_edad_std)
print("Para los que ganan menos de 50k al año, el promedio de edad es: ", menos_50_edad_mean, "y la desviación estándar es: ", menos_50_edad_std)
```

**8. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
# You code here
horas = data['hours-per-week']
max_horas = horas.max()
max_horas_trabajadores = data[data['hours-per-week'] == max_horas]
max_horas_trabajadores_num = len(max_horas_trabajadores)

plus50kmaxhoras = max_horas_trabajadores[max_horas_trabajadores['salary'] == '>50K']
plus50kmaxhoras_count = len(plus50kmaxhoras)

porcentaje = (plus50kmaxhoras_count/max_horas_trabajadores_num)*100
print("La máxima cantidad de horas que alguien trabaja a la semana es de: ", max_horas, "alrededor del ", porcentaje,  " porciento de estos trabajadores ganan más de 50k")
```

**9. Find the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
# You code here
menos50 = data[data['salary'] == '<=50K']
mas50 = data[data['salary'] == '>50K']

menos50_promedio_horas = menos50.groupby('native-country')['hours-per-week'].mean()
más50_promedio_horas = mas50.groupby('native-country')['hours-per-week'].mean()

menos50_promedio_horas_japon = menos50_promedio_horas['Japan']
más50_promedio_horas_japon = más50_promedio_horas['Japan']
print("El promedio de horas trabajadas por personas que ganan menos de 50k es de: ", menos50_promedio_horas_japon)
print("El promedio de horas trabajadas por personas que ganan más de 50k es de: ", más50_promedio_horas_japon)
```

__10. Create 2 visualizations (what to plot is up to you) and write the conclusions generated with those plots from the dataset__

```python
# Your code here: PLOT 1
data['salary_numeric'] = data['salary'].apply(lambda x: 0 if x == '<=50K' else 1)

plt.figure(figsize=(5, 5))
plt.scatter(data['salary_numeric'], data['hours-per-week'], alpha=0.7)
plt.title('Salario vs Horas Trabajadas')
plt.xlabel('Salario (0 = <=50K, 1 = >50K)')
plt.ylabel('Horas trabajadas por semana')
plt.show()
```

```python
# Your code here: PLOT 2

tags = ['<=50K', '>50K']
tamaño = [menos50_promedio_horas_japon, más50_promedio_horas_japon]
plt.figure(figsize=(8, 8))
plt.pie(tamaño, labels=tags, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Promedio de Horas Trabajadas por Personas en Japón (Menos y Más de 50K)')
plt.show()
```
