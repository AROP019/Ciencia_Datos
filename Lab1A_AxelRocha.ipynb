```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/chris1610/pbpython/master/data/sales_data_types.csv"
```

```python
datos = pd.read_csv(url)
```

```python
datos
```

```python
datos.columns
```

```python
#Cambiar nombres columnas
datos.rename(columns={"Customer Number":"Customer_Number"}, inplace=True)
datos.rename(columns={"Customer Name":"Customer_Name"}, inplace=True)
datos.rename(columns={"Percent Growth":"Percent_Growth"}, inplace=True)
datos.rename(columns={"Jan Units":"Jan_Units"}, inplace=True)
datos
```

```python
#Originalmente, no podíamos obtener ventas totales de ningún año por el formato, este se corrigió
datos['2016'] = datos['2016'].replace({'\$': '', ',': ''})
datos['2017'] = datos['2017'].replace({'\$': '', ',': ''})

ventas2016 = datos["2016"].sum()
ventas2017 = datos["2017"].sum()

print("Las ventas en el 2016 fueron de: $",ventas2016)
print("Las ventas en el 2017 fueron de: $",ventas2017)
```

```python
datos.info()
```

```python
#Cambiar tipos de datos de columna
datos['Customer_Number'] = datos['Customer_Number'].round().astype(int)
datos['Customer_Name'] = datos['Customer_Name'].astype(str)
datos['Percent_Growth'] = datos['Percent_Growth'].replace({'%': '', ',': ''}, regex=True).astype(float) / 100.0
datos['Jan_Units'] = datos['Jan_Units'].replace({'Closed': None}, regex=False)
datos['Jan_Units'] = datos['Jan_Units'].round().astype(float)
datos['Active'] = datos['Active'].astype(bool)
datos['Start_Date'] = datos['Month'].astype(str) + '/' + datos['Day'].astype(str) + '/' + datos['Year'].astype(str)
datos['Start_Date'] = pd.to_datetime(datos['Start_Date'], format='%m/%d/%Y')

#Display first 3 rows
display(datos.head(3))
print("")
print("")
print("")

#display first 2 columns and rows with indexes 1 and 2
display(datos.iloc[0:2, :2])
```

```python
# Sort increasingly by 2016 values
datos2 = datos.sort_values(by='2016', inplace=False)
display(datos2)
#Calculate mean of 2016
print("The mean of the values of the 2016 column is: ",datos2['2016'].mean())
```
